```python
from transformers import pipeline
```

```python
# Initialize the summarization model
summarizer = pipeline("summarization")
```

```python
text = """
The quick brown fox jumps over the lazy dog. This sentence has all the letters of the alphabet.
It is commonly used to test typewriters and computer keyboards.
The sentence is a pangram, meaning it contains every letter of the English alphabet.
"""

# Generate a summary
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)

# Print the summary
print(summary)
```
